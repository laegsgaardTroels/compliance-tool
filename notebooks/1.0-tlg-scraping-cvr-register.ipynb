# Wrapping up

```python
import requests
import xml.etree.ElementTree as ET 


def get_root(url):
    return ET.fromstring(
        requests.get(
            url,
            headers={
                'Content-Type': 'application/xml',
            }
        ).content
    )

def get_leaves(node):
    d = {}
    if node:
        for child in node:
            d.update(get_leafs(child))
    else:
        return {node.tag: node.text}
    return d

get_leafs(root)

def parse(hits):
    for hit in hits:
        response = {}
        _source = hit['_source']
        response['cvrNummer'] = _source['cvrNummer']
        for document in _source['dokumenter']:
            if (
                document['dokumentType'] == 'AARSRAPPORT' 
                and document['dokumentMimeType'] == 'application/xml'
            ):
                root = get_root(document['dokumentUrl'])
                response.update(get_leaves(root))
        yield response
        
hits = response['hits']['hits']
responses = []
for idx, r in enumerate(parse(hits)):
    print(f"Parsing {idx}")
    responses.append(r)
```

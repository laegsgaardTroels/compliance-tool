```python
import requests
from bs4 import BeautifulSoup
from functools import reduce


query = "175 darkes lane potters bar money laundering"
query = query.replace(' ', '+')
URL = f"https://google.com/search?q={query}"

# desktop user-agent
USER_AGENT = "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:65.0) Gecko/20100101 Firefox/65.0"
```

```python
headers = {"user-agent" : USER_AGENT}

responses = []
for page in range(991):
    responses.append(
        requests.get(
            URL,
            headers=headers,
            params={'start': page, 'num': page + 10}
        )
    )
```

```python
def get_results(response):
    soup = BeautifulSoup(response.content, "html.parser")
    results = []
    for g in soup.find_all('div', class_='rc'):
        link = g.find_all('a')[0]['href']
        title = g.find('h3').text
        date = g.find_all('span', class_='f')
        if date:
            date = date[0].text.replace('-', '').strip()
        else:
            date = None
        results.append({
            'link': link,
            'title': title,
            'date': date,
        })
    return results
```

```python
results = reduce(
    list.__add__,
    map(
        get_results, 
        filter(
            lambda response: response.status_code == 200,
            responses,
        )
    )
)
```
